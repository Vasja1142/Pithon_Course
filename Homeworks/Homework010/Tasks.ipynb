Задача 1. Найдите, какая область центрального федерального округа имела наибольшую численность студентов вечерней формы обучения в 2015 году.

````python
import pandas as pd
import seaborn as sns
import google.colab as files
````

````python
df = pd.read_csv('https://gbcdn.mrgcdn.ru/uploads/asset/5291776/attachment/817da6a75fe672f89064b6f2bb241a22.csv', sep=';', encoding='cp1251')
df
````

````python
df_1 = df[2:19]
max_value = df_1['Численность студентов очно-заочная (вечерняя) форма, человек, 2015'].max()
df_1 = pd.DataFrame(df_1, columns= ['Субъект РФ','Численность студентов очно-заочная (вечерняя) форма, человек, 2015'])
df_1[df_1['Численность студентов очно-заочная (вечерняя) форма, человек, 2015'] == max_value]
````

Задача 2. Постройте диаграмму с данными о численности студентов дневной формы обучения южного федерального округа за 2017 год.

````python
df_2 = df[33:41]
plot = sns.barplot(df_2, x = 'Субъект РФ', y = 'Численность студентов, очная форма, человек, 2017' )
plot.set_xticklabels(labels = df_2['Субъект РФ'], rotation = 90)
````



Задача 3. Постройте диаграмму количества студентов заочной формы обучения за 2019 год по всем регионам, в которых общее количество студентов не превышает 10000 за данный год.



````python
df_3 = df[df['Численность студентов заочная форма, человек, 2019'] <= 10000]
msk = df_3['Субъект РФ'].str.contains('округ')
df_3 = df_3.drop(df_3[msk].index)
plot = sns.barplot(df_3, x = 'Субъект РФ', y = 'Численность студентов заочная форма, человек, 2019' )
plot.set_xticklabels(labels = df_3['Субъект РФ'], rotation = 90)
````
